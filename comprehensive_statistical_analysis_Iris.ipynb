# Correlation Analysis

```python
correlation_matrix = data.corr(numeric_only=True)
```

```python
correlation_matrix
```

values in the non-diagonal cells represent correlation coefficients, usually ranging from -1 to 1

Here's how to interpret them:
A correlation coefficient of 1 indicates a perfect positive linear relationship. When one variable increases, the other variable increases proportionally.

 A correlation coefficient of 0 means no linear relationship between the variables.
-1: A correlation coefficient of -1 indicates a perfect negative linear relationship. When one variable increases, the other variable decreases proportionally.

Values between -1 and 1 indicate the strength and direction of the linear relationship. The closer the value is to -1 or 1, the stronger the relationship. The sign (+/-) indicates the direction of the relationship: positive or negative, respectively.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation of Features")
plt.show()
```

# Hypothesis Testing

```python
df
```

```python
df = df.iloc[:,[1,2,3,4,5]]
```

```python
df['Species'].unique() # 3 classes i need to figurou the
```

```python
# sepal length and setosa and Iris versicolor
setosa_sepal_length     = df[df['Species'] == 'Iris-setosa']['Sepal Length (cm)']
versicolor_sepal_length = df[df['Species'] == 'Iris-versicolor']['Sepal Length (cm)']
```

```python
t_statistic, p_value = stats.ttest_ind(setosa_sepal_length, versicolor_sepal_length)

#  significance level (alpha)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in sepal length between Iris setosa and Iris versicolor.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in sepal length between Iris setosa and Iris versicolor.")
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Species'], y=df['Sepal Length (cm)'], data=df)
plt.title("Sepal Length Comparison: Iris setosa vs. Iris versicolor")
```
